# DataFrame 조작하기

### DataFrame 생성하기

```python
# iris data를 Pandas의 DataFrame으로 표현
iris_df = pd.DataFrame(load_iris()["data"], columns=load_iris()["feature_names"])
iris_df
```

### DataFrame에 새로운 Column 추가하기

```python
# iris DataFrame에 어떤 꽃인지 나타내는 target column을 추가한다.
iris_df["label"] = load_iris()["target"]
iris_df
```

## DataFrame의 Column에 대하여 조건에 만족하는 문자로 바꾸기

```python
# iris DataFrame에 추가한 label 컬럼을 원본인 문자로 바꿔보자.
target_arr = iris_df["label"].values
target_arr
```

```python
target_names = load_iris()["target_names"]
target_names
```

```python
# 0인 값은 target_name에서 0에 위치하는 값으로 바꾼다.
# 1인 값은 target_name에서 1에 위치하는 값으로 바꾼다.
# 그 외 값은 target_name에서 2에 위치하는 값으로 바꾼다.
iris_df["label"] = np.where(
    target_arr == 0, target_names[0], np.where(
        target_arr == 1, target_names[1], target_names[2]
    )
)
iris_df
```

```python
# 설정된 Index 초기화
iris_df.reset_index(drop=False, inplace=True)
iris_df
```

## 파생 변수 생성하기

```python
# 꽃받침 너비가 3 미만인 경우, "3 미만"으로, 3 이상인 경우 "3 이상"으로, "꽃받침_너비_범주"라는 변수를 생성해보자.
iris_df["꽃받침_너비_범주"] = np.where(iris_df["꽃받침_너비"] >= 3, "3 이상", "3 미만")
iris_df
```

## 빈도분석하기

```python
# "붓꽃_이름"의 빈도표를 생성해보자.
iris_df["붓꽃_이름"].value_counts()
```

```python
# 빈도와 비율을 함께 제공하는 빈도표를 생성해보자.
def Calculate_Freq_table(df, column):
    
    freq_df = pd.DataFrame(df[column].value_counts())
    freq_df = freq_df.reset_index(drop=False).rename(columns={"index":column, column:"Freq"})
    freq_df["Ratio"] = np.round(freq_df["Freq"]/len(df), 2)
    
    return freq_df
```

```python
# "붓꽃_이름"의 빈도와 비율을 계산해보자.
Calculate_Freq_table(iris_df, "붓꽃_이름")
```

```python
# "꽃받침_너비_범주"의 빈도와 비율을 계산해보자.
Calculate_Freq_table(iris_df, "꽃받침_너비_범주")
```

## Multi Index에 대한 빈도표 생성

```python
# "붓꽃_이름"과 "꽃받침_너비_범주"의 빈도표를 생성해보자.
pd.DataFrame(
    iris_df[["붓꽃_이름", "꽃받침_너비_범주"]].value_counts().sort_index(), columns=["Freq"]
)
```

## 연속형 데이터의 전처리
### Min-max Scaler

```python
# 연속형 변수인 연령에 대하여 최솟값과 최댓값을 구하고, 이를 이용해 Min-max scaling하자.
Age_min = titanic_df["Age"].min()
Age_max = titanic_df["Age"].max()

Scaled_Age = (titanic_df["Age"] - Age_min) / (Age_max - Age_min)
```

## Standard Scaler

```python
# Titanic 데이터의 Fare을 Standard Scaling해보자.
Fare_mean = titanic_df["Fare"].mean()
Fare_std = titanic_df["Fare"].std()

Scaled_Fare = (titanic_df["Fare"] - Fare_mean) / Fare_std
```

# Python을 이용한 Tensor 생성
## Titanic 데이터 Tensor화

```python
# 분석에 사용할 Feature만 추출한다.
변수_선택 = titanic_df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
변수_선택
```

```python
# 결측값 제거
결측값_제거 = 변수_선택.dropna().reset_index(drop=True)
결측값_제거
```

## Feature Scaling

```python
# 범주형 변수의 One-Hot Encoding
결측값_제거[["Pclass_1", "Pclass_2", "Pclass_3"]] = pd.get_dummies(결측값_제거["Pclass"])
결측값_제거[["Sex_female", "Sex_male"]] = pd.get_dummies(결측값_제거["Sex"])
결측값_제거[["Embarked_C", "Embarked_Q", "Embarked_S"]] = pd.get_dummies(결측값_제거["Embarked"])
결측값_제거
```

```python
# 연속형 변수를 표준 정규분포로 만드는 함수 생성
def Standard_Scaler(target_Series):
    """
    대상 Series를 Standard Scaler한다.
    """
    target_mean = target_Series.mean()
    target_std = target_Series.std()
    scaled = (target_Series - target_mean)/target_std
    
    return scaled
```

```python
결측값_제거["Age_Scaled"] = Standard_Scaler(결측값_제거["Age"])
결측값_제거["SibSp_Scaled"] = Standard_Scaler(결측값_제거["SibSp"])
결측값_제거["Parch_Scaled"] = Standard_Scaler(결측값_제거["Parch"])
결측값_제거["Fare_Scaled"] = Standard_Scaler(결측값_제거["Fare"])
```

## 전처리된 Feature만 선택

```python
Feature_Scaling = 결측값_제거[
    ["Survived", "Pclass_1", "Pclass_2", "Pclass_3", "Sex_female",
     "Sex_male", "Embarked_C", "Embarked_Q", "Embarked_S", "Age_Scaled",
     "SibSp_Scaled", "Parch_Scaled", "Fare_Scaled"]
]
Feature_Scaling
```

## 생성된 데이터를 Array로 가지고 온다.

```python
# Label을 Array로 가지고 온다.
Train_Label = Feature_Scaling["Survived"].values
Train_Label
```

```python
# Survived만 제외하고 가지고 온다.
Train_Features = Feature_Scaling[Feature_Scaling.columns.difference(["Survived"])].values
Train_Features
```

## 위 과정을 test.csv에 대해서도 해주자.
* test.csv는 Survived 변수를 gender_submission.csv 파일에 입력 해놓았으므로, 이를 합쳐주는 과정을 추가하도록 하자.

## 위에서 실시한 과정을 Test 데이터에도 동일하게 해주자.

```python
# 분석에 사용할 Feature만 추출한다.
test_변수_선택 = test_titanic_df[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]

# 결측값 제거
test_결측값_제거 = test_변수_선택.dropna().reset_index(drop=True)

# 범주형 변수의 One-Hot Encoding
test_결측값_제거[["Pclass_1", "Pclass_2", "Pclass_3"]] = pd.get_dummies(test_결측값_제거["Pclass"])
test_결측값_제거[["Sex_female", "Sex_male"]] = pd.get_dummies(test_결측값_제거["Sex"])
test_결측값_제거[["Embarked_C", "Embarked_Q", "Embarked_S"]] = pd.get_dummies(test_결측값_제거["Embarked"])

# 연속형 변수의 Standard Scaler
test_결측값_제거["Age_Scaled"] = Standard_Scaler(test_결측값_제거["Age"])
test_결측값_제거["SibSp_Scaled"] = Standard_Scaler(test_결측값_제거["SibSp"])
test_결측값_제거["Parch_Scaled"] = Standard_Scaler(test_결측값_제거["Parch"])
test_결측값_제거["Fare_Scaled"] = Standard_Scaler(test_결측값_제거["Fare"])

# 변수 정리
test_Feature_Scaling = test_결측값_제거[
    ["Survived", "Pclass_1", "Pclass_2", "Pclass_3", "Sex_female",
     "Sex_male", "Embarked_C", "Embarked_Q", "Embarked_S", "Age_Scaled",
     "SibSp_Scaled", "Parch_Scaled", "Fare_Scaled"]
]

# Array로 Data를 가지고 온다.
Test_Label = test_Feature_Scaling["Survived"].values
Test_Features = test_Feature_Scaling[test_Feature_Scaling.columns.difference(["Survived"])].values
```

# SVM으로 전처리된 데이터를 학습시켜보자.

```python
from sklearn.svm import SVC

# 선형 SVM으로 학습
clf = SVC(kernel='linear')
clf.fit(Train_Features, Train_Label)
```

## 학습된 결과를 예측 시켜보자.

```python
Predict_result = clf.predict(Test_Features)
Predict_result
```

## 예측된 결과를 혼동행렬로 시각화해보자.

```python
from sklearn import metrics

Confuse_mat = metrics.confusion_matrix(Test_Label, Predict_result)
metrics.ConfusionMatrixDisplay(Confuse_mat).plot()
```
