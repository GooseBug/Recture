# DataFrame 조작하기

### DataFrame 생성하기

```python
# iris data를 Pandas의 DataFrame으로 표현
iris_df = pd.DataFrame(load_iris()["data"], columns=load_iris()["feature_names"])
iris_df
```

### DataFrame에 새로운 Column 추가하기

```python
# iris DataFrame에 어떤 꽃인지 나타내는 target column을 추가한다.
iris_df["label"] = load_iris()["target"]
iris_df
```

## DataFrame의 Column에 대하여 조건에 만족하는 문자로 바꾸기

```python
# iris DataFrame에 추가한 label 컬럼을 원본인 문자로 바꿔보자.
target_arr = iris_df["label"].values
target_arr
```

```python
target_names = load_iris()["target_names"]
target_names
```

```python
# 0인 값은 target_name에서 0에 위치하는 값으로 바꾼다.
# 1인 값은 target_name에서 1에 위치하는 값으로 바꾼다.
# 그 외 값은 target_name에서 2에 위치하는 값으로 바꾼다.
iris_df["label"] = np.where(
    target_arr == 0, target_names[0], np.where(
        target_arr == 1, target_names[1], target_names[2]
    )
)
iris_df
```

## Column의 이름 바꾸기

```python
# DataFrame Column의 이름을 출력해보자.
iris_df.columns
```

```python
# irid_df의 Column 이름들을 바꿔보자.
iris_df.columns = ["꽃받침 길이", "꽃받침 너비", "꽃잎 길이", "꽃잎 너비", "붓꽃"]
iris_df
```

```python
# 특정 Column의 이름만 바꿔보자.
iris_df.rename(columns={"붓꽃":"붓꽃 이름"}, inplace=True)
iris_df
```

```python
# 모든 column의 이름에 있는 " "을 "_"로 바꾸자.
iris_df.columns = iris_df.columns.str.replace(" ", "_")
iris_df
```

## 원하는 Column만 가지고 오기

```python
# 하나의 Column에 대해 Series Type으로 가지고 온다.
iris_df["꽃받침_길이"]
```

```python
# 여러 개의 Column에 대해 DataFrame Type으로 가지고 온다.
iris_df[["꽃받침_길이", "꽃잎_길이", "붓꽃_이름"]]
```

## Index 조작

```python
# 원하는 Column을 Index로 설정
iris_df.set_index("붓꽃_이름", inplace=True)
iris_df
```

```python
# 설정된 Index 초기화
iris_df.reset_index(drop=False, inplace=True)
iris_df
```

## 파생 변수 생성하기

```python
# 꽃받침 너비가 3 미만인 경우, "3 미만"으로, 3 이상인 경우 "3 이상"으로, "꽃받침_너비_범주"라는 변수를 생성해보자.
iris_df["꽃받침_너비_범주"] = np.where(iris_df["꽃받침_너비"] >= 3, "3 이상", "3 미만")
iris_df
```

## 빈도분석하기

```python
# "붓꽃_이름"의 빈도표를 생성해보자.
iris_df["붓꽃_이름"].value_counts()
```

```python
# 빈도와 비율을 함께 제공하는 빈도표를 생성해보자.
def Calculate_Freq_table(df, column):
    
    freq_df = pd.DataFrame(df[column].value_counts())
    freq_df = freq_df.reset_index(drop=False).rename(columns={"index":column, column:"Freq"})
    freq_df["Ratio"] = np.round(freq_df["Freq"]/len(df), 2)
    
    return freq_df
```

```python
# "붓꽃_이름"의 빈도와 비율을 계산해보자.
Calculate_Freq_table(iris_df, "붓꽃_이름")
```

```python
# "꽃받침_너비_범주"의 빈도와 비율을 계산해보자.
Calculate_Freq_table(iris_df, "꽃받침_너비_범주")
```

## Multi Index에 대한 빈도표 생성

```python
# "붓꽃_이름"과 "꽃받침_너비_범주"의 빈도표를 생성해보자.
pd.DataFrame(
    iris_df[["붓꽃_이름", "꽃받침_너비_범주"]].value_counts().sort_index(), columns=["Freq"]
)
```

## csv 파일을 Pandas로 읽어오기

```python
titanic_df = pd.read_csv("train.csv")
titanic_df
```

## 범주형 변수의 Dummy Variable 생성

```python
# 성별을 Dummy 변수로 생성
pd.get_dummies(titanic_df["Sex"])
```

```python
# 성별을 Dummy 변수로 생성하되 맨 뒤의 집단을 기준으로 한다.
pd.get_dummies(titanic_df["Sex"], drop_first=True)
```

## 연속형 데이터의 전처리
### Min-max Scaler

```python
# 연속형 변수인 연령에 대하여 최솟값과 최댓값을 구하고, 이를 이용해 Min-max scaling하자.
Age_min = titanic_df["Age"].min()
Age_max = titanic_df["Age"].max()

Scaled_Age = (titanic_df["Age"] - Age_min) / (Age_max - Age_min)
```

```python
# 표준화 전과 후를 시각화하여 보자.
import matplotlib.pyplot as plt
```

```python
plt.hist(titanic_df["Age"])
plt.show()
```

```python
plt.hist(Scaled_Age)
plt.show()
```

## Standard Scaler

```python
# Titanic 데이터의 Fare을 Standard Scaling해보자.
Fare_mean = titanic_df["Fare"].mean()
Fare_std = titanic_df["Fare"].std()

Scaled_Fare = (titanic_df["Fare"] - Fare_mean) / Fare_std
```

```python
# 표준화 전 시각화
plt.hist(titanic_df["Fare"])
plt.show()
```

```python
# 표준화 후 시각화
plt.hist(Scaled_Fare)
plt.show()
```
